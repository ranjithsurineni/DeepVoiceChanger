```python
import numpy as np
import matplotlib.pyplot as plt
import os

# Set the path to one of the processed .npy files
processed_file = "D:\\projects\\voice_conversion\\data\\processed\\LJ001-0001.npy"

# Load the .npy file
spectrogram = np.load(processed_file)

# Plot the spectrogram
plt.figure(figsize=(10, 4))
plt.imshow(spectrogram, aspect='auto', origin='lower')
plt.colorbar(label="Decibels")
plt.title("Spectrogram of LJ001-0001.wav")
plt.xlabel("Time Frames")
plt.ylabel("Mel Frequency Bins")
plt.show()
```
