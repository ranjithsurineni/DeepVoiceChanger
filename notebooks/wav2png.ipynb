```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt

#  Step 1: Load the WAV file
file_path = "D:\\projects\\voice_conversion\\data\\LJSpeech-1.1\\wavs\\LJ001-0001.wav"  # file path
y, sr = librosa.load(file_path, sr=None)  # Load with original sampling rate

#  Step 2: Plot the waveform
plt.figure(figsize=(12, 4))
librosa.display.waveshow(y, sr=sr, alpha=0.8)
plt.title("Waveform of the Audio File")
plt.xlabel("Time (seconds)")
plt.ylabel("Amplitude")
plt.show()

#  Step 3: Print Audio Info
duration = librosa.get_duration(y=y, sr=sr)
print(f" Audio Loaded: {file_path}")
print(f" Sample Rate: {sr} Hz")
print(f" Duration: {duration:.2f} seconds")

#  Step 4: Load and Plot the Spectrogram
processed_file = "D:\\projects\\voice_conversion\\data\\processed\\LJ001-0001.npy"  # Replace with actual .npy file path
spectrogram = np.load(processed_file)  # Load preprocessed spectrogram

plt.figure(figsize=(10, 4))
plt.imshow(spectrogram, aspect='auto', origin='lower', cmap='inferno')
plt.colorbar(label="Decibels")
plt.title("Spectrogram of Processed Audio")
plt.xlabel("Time Frames")
plt.ylabel("Mel Frequency Bins")
plt.show()
```
